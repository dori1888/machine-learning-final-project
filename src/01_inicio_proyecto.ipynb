```python
import os

os.listdir("data/raw/IHME_GBD_2021_DEMOG_1950_2021_TABLES")
```

```python
import pandas as pd

file_path = (
    "data/raw/"
    "IHME_GBD_2021_DEMOG_1950_2021_TABLES/"
    "IHME_GBD_2021_DEMOG_1950_2021_TABLE_1_Y2024M03D11.CSV"
)

df = pd.read_csv(file_path)
df.head()
```

```python
df.columns
df.head()
```

```python
cols = [
    "lancet_label",
    "val.lexboth",
    "val.lexm",
    "val.lexf",
    "val.deaths",
    "val.u5mort"
]

df_clean = df[cols]
df_clean.head()
```

```python
df_clean = df_clean.rename(columns={
    "lancet_label": "region",
    "val.lexboth": "life_expectancy_total",
    "val.lexm": "life_expectancy_male",
    "val.lexf": "life_expectancy_female",
    "val.deaths": "total_deaths",
    "val.u5mort": "under5_mortality"
})

df_clean.head()
```

```python
df_clean.describe()
```

```python
df_clean.sort_values("life_expectancy_total", ascending=False).head(10)
```

```python
df_clean.sort_values("under5_mortality", ascending=False).head(10)
```

```python
bottom10 = df_clean.sort_values(
    "life_expectancy_total"
).head(10)

plt.figure(figsize=(10,6))
plt.barh(
    bottom10["region"],
    bottom10["life_expectancy_total"]
)
plt.xlabel("Años")
plt.title("10 regiones con menor esperanza de vida")
plt.gca().invert_yaxis()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(
    df_clean["under5_mortality"],
    df_clean["life_expectancy_total"]
)

plt.xlabel("Mortalidad infantil (<5)")
plt.ylabel("Esperanza de vida total")
plt.title("Relación entre mortalidad infantil y esperanza de vida")
plt.show()
```

```python
df_final = df_clean[[
    "region",
    "life_expectancy_total",
    "life_expectancy_male",
    "life_expectancy_female",
    "under5_mortality",
    "total_deaths"
]]
```

```python
df_final.head()
df_final.shape
```

```python
top10 = df_final.sort_values(
    "life_expectancy_total",
    ascending=False
).head(10)

import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.barh(
    top10["region"],
    top10["life_expectancy_total"]
)
plt.xlabel("Años")
plt.title("Top 10 regiones con mayor esperanza de vida")
plt.gca().invert_yaxis()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(
    df_final["under5_mortality"],
    df_final["life_expectancy_total"]
)

plt.xlabel("Mortalidad infantil (<5 años)")
plt.ylabel("Esperanza de vida total")
plt.title("Relación entre mortalidad infantil y esperanza de vida")
plt.show()
```

```python
df_final["gap_gender"] = (
    df_final["life_expectancy_female"]
    - df_final["life_expectancy_male"]
)

top_gap = df_final.sort_values(
    "gap_gender",
    ascending=False
).head(10)

plt.figure(figsize=(10,6))
plt.barh(
    top_gap["region"],
    top_gap["gap_gender"]
)
plt.xlabel("Años")
plt.title("Top 10 regiones con mayor diferencia de esperanza de vida (Mujer - Hombre)")
plt.gca().invert_yaxis()
plt.show()
```

### Análisis exploratorio

Los datos muestran una fuerte relación inversa entre la mortalidad infantil y la esperanza de vida total.
Las regiones con menor mortalidad en menores de cinco años presentan sistemáticamente mayores niveles de longevidad.

Además, se observa una diferencia consistente entre la esperanza de vida femenina y masculina,
siendo mayor en mujeres en prácticamente todas las regiones analizadas.
